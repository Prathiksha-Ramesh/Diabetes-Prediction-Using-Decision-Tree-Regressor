## Dieabetes Prediction- Decision tree Regressor 

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
print(dataset['DESCR'])
```

```python
dataset
```

```python
import pandas as pd
df=pd.DataFrame(dataset.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
df.head()
```

```python
X=df
y=dataset['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

## Hyperparameter tuning

```python
params={'criterion':['squared_error','friedman_mse','absolute_error'],
        'splitter':['best','random'],
        'max_depth':[1,2,3,4,5,10,15,20,25],
        'max_features':['auto','sqrt','log2']
        }
```

```python
from sklearn.model_selection import GridSearchCV
regressor=DecisionTreeRegressor()
grid=GridSearchCV(regressor,param_grid=params,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(r2_score(y_pred,y_test))
print(mean_absolute_error(y_pred,y_test))
print(mean_absolute_error(y_pred,y_test))
```

```python
regressor=DecisionTreeRegressor(criterion='squared_error',max_depth=3,max_features='sqrt',splitter= 'random')
```

```python
regressor.fit(X_train,y_train)
```

```python
#visualization

import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(regressor,filled=True)
```
